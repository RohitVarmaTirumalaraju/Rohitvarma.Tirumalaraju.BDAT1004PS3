#### step 9

```python
def stats(var):
    var = pd.Series(var)
    Min = var.min()
    Max = var.max()
    Mean = var.mean()
    Std = var.std()
    results = [Min,Max,Mean,Std]
    indexval = ["Minimun","Maximum","Mean","Standard Deviation"]
    results = pd.Series(results,index=indexval)
    return results
loc_stats = DataNew.apply(stats)
loc_stats
```

## Question 10

Create a scatter plot showing the relationship between the total revenue earned by
arcades and the number of Computer Science PhDs awarded in the U.S. between
2000 and 2009.<br>
Don't forget to label your axes!<br>
Color each dot according to its year.

```python
import pandas as pd
import seaborn as sbl
import matplotlib.pyplot as dataplot
%matplotlib inline
revenue = pd.read_csv('C:\\Users\\Downloads\\arcade-revenue-vs-cs-doctorates.csv', index_col=0)
```

```python
sbl.set(rc = {'figure.figsize':(12,10)})
sbl.set_theme(style="white")
sbl.scatterplot(x='Total Arcade Revenue (billions)', y='Computer Science Doctorates Awarded (US)', 
                hue='Year', data=revenue, palette="deep",s=200)
dataplot.title('CS Doctorates vs Revenue', color = 'Red', fontsize = 18,fontweight ='bold')
dataplot.xlabel('Total Arcade Revenue in billions' , color = 'Blue', fontsize = 14)
dataplot.ylabel('CS Doctorates Awarded', color = 'Blue', fontsize = 14)
```
